# Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Indeed Dataset/indeed_job_dataset.csv")
df.head()
```

```python
df.columns
```

```python
df = df[['Job_Title', 'Queried_Salary', 'Job_Type',
       'Skill', 'No_of_Skills', 'Company', 'No_of_Reviews', 'No_of_Stars',
       'Date_Since_Posted', 'Description', 'Location', 'Company_Revenue',
       'Company_Employees', 'Company_Industry']]
```

```python
df.head()    
```

# Number of Jobs Bar-graph

```python
df.head()
```

```python
job_types = df.groupby(["Company_Industry"])["Job_Type"].value_counts()
job_types = pd.DataFrame(job_types)
job_types.columns = ["Job_Count"]
job_types = job_types.reset_index()
job_types = job_types.set_index("Company_Industry")
job_types.head(15)
```

```python
job_types = job_types.pivot(columns = "Job_Type", values = "Job_Count").fillna(0)
job_types
```

```python
job_types.groupby(["Job_Type"])
```

# By Salary

```python
salary_by_industry = df.groupby(["Company_Industry"])["Queried_Salary"].value_counts()
salary_by_industry = pd.DataFrame(salary_by_industry)
```

```python
salary_by_industry.head(50)
```

```python
salary_by_industry.plot.bar()
```

```python
job_order = job_types.groupby(["Company_Industry"])["Job Count"].sum()
job_order = job_order.sort_values(ascending = False)
job_order.head(15)
```
