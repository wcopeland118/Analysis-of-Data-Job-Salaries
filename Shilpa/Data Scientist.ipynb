```python
filtered_data = data.filter(["Job_Type", "Queried_Salary", "Skill", "No_of_Skills", 
             "Location", "Company_Industry"]) 
filtered_data
```

```python
Two_columns = filtered_data.filter(["Job_Type", "Location"])
Two_columns.dropna()
Two_columns["Location"].fillna("NAN").value_counts().head()
```

```python
Data_Scientist = Two_columns[Two_columns["Job_Type"] == "data_scientist"]

DS = Data_Scientist["Location"].fillna("Not Reported").value_counts()
DS
DS1 = pd.DataFrame(DS)
DS1.columns = ["Data Scientist"]
DS1.head()
```

```python
DS_Final = DS1.drop(DS1[DS1["Data Scientist"] < 10].index)
DS_Final.head()
```

```python
Data_Analyst = Two_columns[Two_columns["Job_Type"] == "data_analyst"]
DA = Data_Analyst["Location"].fillna("Not Reported").value_counts()
```

```python
DA1 = pd.DataFrame(DA)
DA1.columns = ["Data Analyst"]
DA1
```

```python
Data_Engineer = Two_columns[Two_columns["Job_Type"] == "data_engineer"]
DE = Data_Engineer["Location"].fillna("Not Reported").value_counts()
DE1 = pd.DataFrame(DE)
DE1.columns = ["Data Engineer"]
DE1.head()
```

```python
Jobs_Location = DS1.join(DA1).join(DE1)
Jobs_Location.index.name = "States"
Jobs_Location.head()
```
